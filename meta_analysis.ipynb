$\hspace{1cm}$ Meta-analiza to ilościowa synteza wyników pochodzących z niezależnych badań prowadzonych na różnych grupach populacyjnych. W kontekście badań case-control, metaanaliza uważana jest za najbardziej wiarygodną formę analizy wśród badań klinicznych. W ramach metaanalizy nie gromadzi się nowych danych, lecz łączy wyniki już przeprowadzonych badań, zwłaszcza prób klinicznych, które zazwyczaj obejmują niewielką liczbę pacjentów, a uzyskane wyniki często nie pozwalają na jednoznaczne wnioskowanie.

# Wstęp

## Choroba Parkinsona

$\hspace{1cm}$ Choroba Parkinsona (PD) jest chorobą zwyrodnieniową ośrodkowego układu nerwowego (OUN). Objawy ruchowe PD wynikają ze śmierci komórek wytwarzających dopaminę w regionie istoty czarnej śródmózgowia. Chociaż etiologia PD nie jest w pełni poznana, uważa się, że obejmuje ona interakcje między podatnym podłożem genetycznym a czynnikami środowiskowymi.

## Gen VDR

$\hspace{1cm}$ Mimo że główną rolą witaminy D jest utrzymanie homeostazy mineralnej kości, wywiera ona istotny wpływ na komórki odpornościowe i nerwowe jako neuroimmunomodulator. Witamina D wpływa na rozwój oraz utrzymanie neuronów, odgrywając jednocześnie rolę neuroprotekcyjną w centralnym układzie nerwowym (OUN). Jej niedobór związany jest między innymi z chorobą Parkinsona (PD). Ze względu na ochronne działanie witaminy D i potencjalny związek między jej niedoborem a chorobami neurodegeneracyjnymi, przeprowadzono badania polimorfizmów receptora witaminy D (VDR). Receptor ten, kodujący jądrowy czynnik transkrypcyjny, znajduje się na chromosomie 12q13.11. Zidentyfikowano trzy polimorfizmy w silnej nierównowadze sprzężeń (LD) - BsmI (rs1544410) oraz ApaI (rs7975232) (oba w intronie 8), a także TaqI (rs731236) (w eksonie 9) na 3'-końcu genu. Dodatkowo, inny obszar polimorficzny, FokI (rs2228570), zlokalizowany jest w kodonie startowym. Ta zmienność alleliczna może mieć istotne znaczenie funkcjonalne, jednak opublikowane wyniki na ten temat są niespójne.

Przeprowadziliśmy metaanalizę w celu ustalenia, czy polimorfizmy VDR są powiązane z podatnością na PD.

## Użyte metody statystyczne

$\hspace{1cm}$ Do analizy statystycznej wykorzystano następujące metody: **$OR$** (odds-ratio) w celu sprawdzenia czy wystąpienie danego polimorfizmu jest ściśle związane z narażeniem na rozwinięcie PD; **wykres leśny i lejkowy** (forest and funnel plot) w celu podsumowania informacji na temat uwzględnionych badań; **$I^2$** w celu oceny heterogenności tychże badań. Przyjęty poziom isotności wynosił 5% ($\alpha = 0.05$).

# Przebieg Meta-Analizy

## Intstalacja pakietu `meta` i jego wczytanie

```r
install.packages('meta')

library(meta)

# setting the plot sizes
options(repr.plot.width=9, repr.plot.height=4)
```

## Wczytanie danych z reposytorium [GitHub](https://github.com/KacperKaszuba0608/meta-analysis)

```r
# read in apaI csv
apaI <- read.csv('https://github.com/KacperKaszuba0608/meta-analysis/raw/main/apaI.csv', header = TRUE)

# read in bsmI csv
bsmI <- read.csv('https://github.com/KacperKaszuba0608/meta-analysis/raw/main/bsmI.csv', header = TRUE)

# read in fokI csv
fokI <- read.csv('https://raw.githubusercontent.com/KacperKaszuba0608/meta-analysis/main/fokI.csv', header = TRUE)

# read in taqI csv
taqI <- read.csv('https://github.com/KacperKaszuba0608/meta-analysis/raw/main/taqI.csv', header = TRUE)
```

### Wyniki Meta-Analizy

Tabele zbiorcze dla poszczególnych wariantów genu VDR, zawierają następujące kolumny:
* `ID` - numer identyfikacyjny badania
* `Study` - pierwszy autor badania
* `Population` - etniczność
* `Year` - data publikacji badania
* `A` - liczebność homozygot dominujących w grupie chorych na PD
* `N.AC.` - suma liczebności heterozygot i homozygot recesywnych w grupie chorych na PD
* `B` - liczebność homozygot dominujących w grupie kontrolnej
* `N.BD.` - suma liczebności heterozygot i homozygot recesywnych w grupie kontrolnej

Następnie wyniki liczebności z każdej z nich zostały poddane analizie statystycznej za pomocą, której obliczono łączne Odds-Ratio oraz statystykę $I^2$ określająca heterogeniczność.

```r
# @title ApaI
# @markdown Charakterystyka poszczególnych badań objętych meta-analizą
(apaI <- cbind(ID = 1:nrow(apaI), apaI))
# metabin of apaI
meta.apaI <- metabin(A, N.AC., B, N.BD., data = apaI, sm = "OR", method = "I",
                     comb.fixed = F, studlab = Study)
```

$\hspace{1cm}$ Poniższy wykres przedstawia łączne Odds-Ratio dla uzwględnionych badań dotyczących wariatnu *ApaI*. Wynosiło ono $0.79$ $(95\%$ $CI, 0.43 - 1.44)$ co mówi nam, że ten wariant jest związany z narażeniem na rozwinięcie PD. Wartość dotycząca heterogenności ($I^2$) badań wynosi $90\%$ $(p < 0.01)$, co oznacza istotną statystycznie niejednorodność. Jest to równoważne ze stwierdzeniem, iż badania dotyczące owego wariantu mogą być niespójne z innego powodu niż przypadek.

```r
forest(meta.apaI)
```

$\hspace{1cm}$ Poniższy wykres lejkowy obrazuje niski błąd systematyczny publikacji. Większość punktów reprezentujących badania (wpływ *ApaI*) znalazło się w prawidłowym obszarze $(p < 0.05)$. Reprezentuje on wysoką precyzję oraz przedstawia silny wpływ średniego efektu polimorfizmu na PD. Warto zauważyć, że 2 badania, które mają silny wpływ na grupy osób chorych (punkt po lewej) i zdrowych (punkt po prawej) znajdują się poza obszarem krytycznym.

```r
funnel(meta.apaI)
```

```r
# @title BsmI
# @markdown Charakterystyka poszczególnych badań objętych meta-analizą
(bsmI <- cbind(ID = 1:nrow(bsmI), bsmI))
# metabin of bsmI
meta.bsmI <- metabin(A, N.AC., B, N.BD., data = bsmI, sm = "OR", method = "I",
                     comb.fixed = F, studlab = Study)
```

$\hspace{1cm}$ Poniższy wykres przedstawia łączne Odds Ratio dla uzwględnionych badań. Wynosiło ono $0.94$ $(95\%$ $CI, 0.52 - 2.28)$ co mówi nam, że wariant typu *BsmI* jest nieznacznie związany z rozwinięciem PD. Wartość dotycząca heterogenności ($I^2$) badań wynosi $0\%$ $(p = 0.48)$, co oznacza nieistotną statystycznie niejednorodność. Jest to również równoważne ze stwierdzeniem, iż badania dotyczące owego wariantu są spójne.

```r
forest(meta.bsmI)
```

$\hspace{1cm}$ Poniższy wykres lejkowy obrazuje brak błędu systematycznego publikacji. Każdy punkt reprezentujący badanie (wpływ *BsmI*) znalazł się w prawidłowym obszarze $(p < 0.05)$. Reprezentuje on średnią precyzję oraz przedstawia sliny średni efektu polimorfizmu na grupę kontrolną.

```r
funnel(meta.bsmI)
```

```r
# @title FokI
# @markdown Charakterystyka poszczególnych badań objętych meta-analizą
(fokI <- cbind(ID = 1:nrow(fokI), fokI))

# metabin of fokI
meta.fokI <- metabin(A, N.AC., B, N.BD., data = fokI, sm = "OR", method = "I",
                     comb.fixed = F, studlab = Study)
```

$\hspace{1cm}$ Poniższy wykres przedstawia łączne Odds Ratio dla uzwględnionych badań. Wynosiło ono $1.11$ $(95\%$ $CI, 0.86 - 1.44)$ co mówi nam, że wariant typu *FokI* częściej występuje u osób zdrowych. Wartość dotycząca heterogenności ($I^2$) badań wynosi $71\%$ $(p < 0.01)$, co oznacza istotną statystycznie niejednorodność. Jest to również równoważne ze stwierdzeniem, iż badania dotyczące owego wariantu mogą być niespójne z innego powodu niż przypadek.

```r
forest(meta.fokI)
```

$\hspace{1cm}$ Poniższy wykres lejkowy obrazuje niski błąd systematyczny publikacji. Tylko jeden punkt reprezentujący badanie (wpływ *FokI*) nie znalazła się w prawidłowym obszarze $(p < 0.05)$. Reprezentuje on wysoką precyzję oraz przedstawia silny wpływ średniego efektu polimorfizmu na grupę kontrolną. Tylko jedno badanie miało bardzo niską precyzję.

```r
funnel(meta.fokI)
```

```r
# @title TaqI
# @markdown Charakterystyka poszczególnych badań objętych meta-analizą
(taqI <- cbind(ID = 1:nrow(taqI), taqI))
# metabin of taqI
meta.taqI <- metabin(A, N.AC., B, N.BD., data = taqI, sm = "OR", method = "I",
                     comb.fixed = F, studlab = Study)
```

$\hspace{1cm}$ Poniższy wykres przedstawia łączne Odds Ratio dla uzwględnionych badań. Wynosiło ono $0.84$ $(95\%$ $CI, 0.73 - 0.96)$ co mówi nam, że wariant typu *TaqI* jest silniej związany z rozwinięciem PD. Wartość dotycząca heterogenności ($I^2$) badań wynosi $0\%$ $(p = 0.56)$, co oznacza nieistotną statystycznie niejednorodność. Jest to również równoważne ze stwierdzeniem, iż badania dotyczące owego wariantu są spójne.

```r
forest(meta.taqI)
```

$\hspace{1cm}$ Poniższy wykres lejkowy obrazuje brak błędu systematycznego publikacji. Każdy punkt reprezentujący badanie (wpływ *TaqI*) znalazł się w prawidłowym obszarze $(p < 0.05)$. Reprezentuje on średnią precyzję oraz przedstawia umiarkowany wpływ średniego efektu polimorfizmu na chorobą Parkinson'a. Tylko jedno z badań miało niską precyzję.

```r
funnel(meta.taqI)
```

# Podsumowanie

$\hspace{1cm}$ Podsumowując uzyskane wyniki, nie można stwierdzić, że wszystkie warianty odpowiadają za narażenie rozwój choroby Parkinson'a. Okazało się, że łączne $OR$ dla polimorfizmu *FokI* wskazuje grupę kontrolną jako tą, w której genie VDR pojawia się częściej wariant heterozygotyczny lub homozygotyczny recesywny. Pozostałe warianty potwierdzają założenie związku z PD. Jednakże w wariantach *ApaI* oraz *FokI* przedziały ufności, w których znajdują się wartości $OR$ są długie i przechodzą przez efekt zerowy (brak związku z żadną grupą - 1). Świadczy to o tym, że nie są to statystycznie istotne wyniki. Dlatego też jedynie wariant TaqI potwierdza hipotezę, że ma on związek z rozwojem jednostki chorobowej PD.

$\hspace{1cm}$ Analizując wyniki wskazujące zróżnicowanie rezultatów ($I^2$), można stwierdzić, że warianty *ApaI* i *FokI* posiadają statystycznie istotną heterogeniczność, natomiast warianty *BsmI* oraz *TaqI* posiadają nieistotną heterogeniczność. Oznacza to, że wyniki dla wariantów statystycznie istotnych są bardzo zróżnicowane, podczas gdy wyniki dla wariantów nieistotnych są podobne względem siebie.

## ApaI

$\hspace{1cm}$ Zagłębiając się w wartości OR dla poszczególnych badań, warto zauważyć, że dla wariantu *ApaI* szczególnie wyróżniają się 2 grupy etniczne pochodzące z Iranu. Mianowicie są to Irańczycy z całego kraju oraz Irańczycy z południowo-zachodniej części. W przypadku pierwszej grupy iloraz szans wyniósł 7.93 z bardzo długim 95% przedziałem ufności. Natomiast drugie badanie wykazało, że Południowo-Zachodni Irańczycy są bardziej narażeni na ten wariant (OR = 0.08 z krótim 95% CI). Długości przedziałów ufności prawdopodobnie wynikają z ilości osób biorących udział w badaniu, gdyż w pierwszej wspomnianej populacji całkowita liczba osób wynosiła zaledwie 112. Dla porównania druga publikacja miała 210 osób.

## BsmI

$\hspace{1cm}$ W przypadku wariantu BsmI wyróżnić można 3 grupy etniczne, dla których wspomniany polimorfizm może mieć związek z rozwojem choroby Parkinson'a. Są to Kalifornijczycy, Kaukascy Włosi oraz Japończycy. Co ciekawe współczynnik ilorazu szans był statystycznie istotny jedynie dla Kalifronijczyków (OR = 0.7, 95% CI, 0.51 - 0.98). Pozostałe grupy miały OR niższe niż 1 ale ich przedziały ufności przecinały efekt zerowy.

## FokI

$\hspace{1cm}$ Iloraz sznas, który wskazywałby że wariant *FokI* ma związek z rozowjem PD dotyczył jedynie Kaukaskich Włochów (OR = 0.61, 95% CI, 0.48 - 0.78). Pozostałe rezultaty nie potwierdzają założenia, że wariant *FokI* powoduje chorobę Parkinson'a.

## TaqI

$\hspace{1cm}$ Meta-analiza wariantu *TaqI* wykazała, że jest on ściśle związany z rozwojem jednostki chorobowej Parkinson'a. Jednakże w przypadku 2 grup iloraz szans wskazał ten wariant jako częstszy u grupy kontrolnej. Były to osoby z Wysp Owczych oraz Iranu. Okazuje się, że są to nieistotne statystycznie wyniki, ponieważ ich 95% przedziały ufności przecinają efekt zerowy, więc nie można jednoznacznie stwierdzić, czy jest to prawdą. Mimo znacznej przewagi rezultatów optujących za tym, że *TaqI* ma istotny związek z PD, to jedynie dla grupy z Kaliforni obliczony iloraz szans był statystycznie isotny.
